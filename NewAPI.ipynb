```python
import requests
import json
query = input("What Type of News Are You Intrested In?")

url = f"https://newsapi.org/v2/everything?q={query}&from=2024-07-29&to=2024-07-29&sortBy=popularity&apiKey=da392557ba374928bd37f8e3eef9c096"

r = requests.get(url)
news = json.loads(r.text)
# print(news, type(news))

for article in news ["articles"]:
    print(article["title"])
    print(article["description"])
    print(article["url"])
    print("-"*20)
```
